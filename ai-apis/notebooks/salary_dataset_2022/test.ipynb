```python
import pickle

path = "../../pre_trained/salary_2022/model.pkl"

with open(path, "rb") as f:
  model = pickle.load(f)

with open("../../pre_trained/salary_2022/scaler.pkl", "rb") as f:
  scaler = pickle.load(f)

model
scaler
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("../../data/Salary_Dataset_with_Extra_Features.csv")
```

```python
data = pd.DataFrame({
  "company_name" : data["Company Name"],
  "job_roles" : data["Job Roles"],
  "location" : data["Location"],
  "salary" : data["Salary"],
  "employment_status" : data["Employment Status"]
})

labels = ["company_name" , "job_roles" , "location", "employment_status"]
data = pd.get_dummies(data=data, columns=labels)
ip = data.drop(columns=["salary"])

ip

# model.predict([[company_name, job_roles, location]])
```

```python
company_name = "IBM"
job_roles = "Web"
location = "Bangalore"
employment_status = "Full Time"

new_data = pd.DataFrame({
    "company_name": [company_name],
    "job_roles": [job_roles],
    "location": [location],
    "employment_status" : [employment_status]
})

new_data_encoded = pd.get_dummies(new_data, columns=["company_name", "job_roles", "location", "employment_status"])
new_data_encoded
```

```python
new_data_encoded = new_data_encoded.reindex(columns=ip.columns, fill_value=0)
op = model.predict(new_data_encoded)

op = scaler.inverse_transform(op.reshape(-1, 1)) #-1 means row take automatically as ip

op[0][0]
```
