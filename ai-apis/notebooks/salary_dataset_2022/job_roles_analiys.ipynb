```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../../data/Salary_Dataset_with_Extra_Features.csv")
data.head(2)
```

### Job Roles Analysis

1 JOb roles graph

```python
roles = data.groupby(by="Job Roles")["Employment Status"].count().sort_values(ascending=False).reset_index()
roles
```

```python
bar = plt.bar(roles["Job Roles"], roles["Employment Status"])
plt.bar_label(bar, labels=roles["Employment Status"])
plt.xlabel('Roles');
plt.ylabel('Count');
plt.title('Job roles applied');
plt.xticks(rotation=45)
plt.show()
```

avg salary by roles

```python
avg_salary = data.groupby(by="Job Roles")["Salary"].mean().round(0).reset_index()
avg_salary
```

```python
bar = plt.bar(avg_salary["Job Roles"], avg_salary["Salary"])
plt.xticks(rotation=45)
plt.xlabel("Roles")
plt.ylabel("Avg salary")
plt.title("Avg salary by Roles")

plt.show()
```

location wise count of company

```python
company_cnt = data.groupby(by="Location")["Company Name"].count().reset_index()
company_cnt
```

```python
bars = plt.bar(company_cnt["Location"], company_cnt["Company Name"])
plt.bar_label(bars, labels=company_cnt["Company Name"])
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Company Count")
plt.title("Company Distribution by Location")
plt.show()
```

location wise salary

```python
salary_loc = data.groupby(by="Location")["Salary"].mean().round()
salary_loc
```

```python
plt.bar(salary_loc["Location"], salary_loc["Salary"])
plt.xlabel("Location")
plt.ylabel("Avg salary")
plt.xticks(rotation=45)
plt.title("Location wise salary")
plt.show()
```

company rating with location

```python
rating = data.groupby(by="Location")["Rating"].mean().reset_index()
rating
```

```python
plt.bar(rating["Location"], rating["Rating"])
plt.xticks(rotation=45)
plt.show()
```
